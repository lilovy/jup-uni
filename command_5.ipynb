### Функция построения графиков: экспериментальный, теоретический

```python
def build_graphic(list_time, list_count_elements, n_change):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.legend('fs')
    plt.plot(list_count_elements, list_time, color = 'blue', label = 'экспериментальный')
    ax2 = ax.twinx()
    plt.plot(list_count_elements, n_change, color = 'orange', label = 'теоретический ')
    fig.legend()
    plt.show()
```

### 1-й алгоритм. Сортировка

```python
def sort_array(array, list_time, list_count_elements):
    time_start = time.perf_counter()
    for i in range(len(array)):
        for j in range(len(array) - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
    list_time.append(time.perf_counter() - time_start)
    list_count_elements.append(len(array))
```

```python
# Вызов функции алгоритма сортировки
list_time = []
list_count_elements = []
for n in n_list:
    array = [randint(1,15) for _ in range(n)]
    sort_array(array, list_time, list_count_elements)
```

### Алгоритм сортировки. Построение графика: экспериментальный, теоретический 

```python
n_power_2 = []
for n in n_list:
    n_power_2.append(n**2)
```

```python
build_graphic(list_time, list_count_elements, n_power_2)
```

### 2-й алгоритм. Подсчет количество уникальных элементов

```python
def count_unique_elements(array, list_time, list_count_elements):
    time_start = time.perf_counter()
    new_array = []
    count = 0
    for elem in array:
        if elem not in new_array:
            new_array.append(elem)
            count += 1
    list_time.append(time.perf_counter() - time_start)
    list_count_elements.append(len(array))
```

```python
# Вызов функции алгоритма подсчета количества уникальных элементов
list_time = []
list_count_elements = []
for n in n_list:
    array = [randint(1,15) for _ in range(n)]
    count_unique_elements(array, list_time, list_count_elements)
```

### Алгоритм подсчета количества уникальных элементов. Построение графика: экспериментальный, теоретический

```python
build_graphic(list_time, list_count_elements, list_count_elements)
```

### 3-й алгоритм. Поиск нужного элемента

```python
def search_desired_item(array, list_time, list_count_elements):
    time_start = time.perf_counter()
    n = 5
    for i in range(len(array)):
        if n == i:
            list_time.append(time.perf_counter() - time_start)
            list_count_elements.append(len(array))
```

```python
# Вызов функции алгорима поиска нужного элемента
list_time = []
list_count_elements = []
for n in n_list:
    array = [randint(1,15) for _ in range(n)]
    search_desired_item(array, list_time, list_count_elements)
```

### Алгоритм поиска нужного элемента. Построение графика: экспериментальный, теоретический

```python
build_graphic(list_time, list_count_elements, list_count_elements)
```
